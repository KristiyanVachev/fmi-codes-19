### Pickling

```python
import _pickle as cPickle
from pathlib import Path

def dumpPickle(fileName, content):
    pickleFile = open(fileName, 'wb')
    cPickle.dump(content, pickleFile, -1)
    pickleFile.close()

def loadPickle(fileName):    
    file = open(fileName, 'rb')
    content = cPickle.load(file)
    file.close()
    
    return content
    
def pickleExists(fileName):
    file = Path(fileName)
    
    if file.is_file():
        return True
    
    return False
```

```python
trainPickleName = 'pickles/trainWordsDf.pkl'
train = loadPickle(trainPickleName)

devPickleName = 'pickles/devWordsDf.pkl'
dev = loadPickle(devPickleName)
```

```python
def oneHotEncodeColumns(df):
    columnsToEncode = ['NER', 'POS', "TAG", 'DEP']

    for column in columnsToEncode:
        print(column)
        one_hot = pd.get_dummies(df[column])
        one_hot = one_hot.add_prefix(column + '_')

        df = df.drop(column, axis = 1)
        df = df.join(one_hot)
    
    return df
```

```python
train = oneHotEncodeColumns(train)
```

```python
train.head()
```

```python
dev = oneHotEncodeColumns(dev)
```

```python
dev.head()
```

```python
columnsToDrop = ['text', 'titleId', 'paragrapghId', 'sentenceId', 'shape']
```

```python
train = train.drop(columnsToDrop, axis = 1)
dev = dev.drop(columnsToDrop, axis = 1)
```

```python
train.head()
```

```python
dev.head()
```

```python
#Both dataframe's features
features = []

for feature in dev.columns:
    if feature not in features:
        features.append(feature)
        
for feature in train.columns:
    if feature not in features:
        features.append(feature)
```

```python
# Add missing features as columns
for feature in features:
    if feature not in train.columns:
        dev[feature] = 0

for feature in features:
    if feature not in dev.columns:
        dev[feature] = 0
```

```python
x_train = train.drop(labels=['isAnswer'], axis=1)
y_train = train['isAnswer']

x_test = dev.drop(labels=['isAnswer'], axis=1)
y_test = dev['isAnswer']
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
y_test.head()
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
```

```python
predictor = gnb.fit(x_train, y_train)
y_pred = predictor.predict(x_test)
```

```python
correctCount = (y_test == y_pred).sum()

print('Correctly guessed:', '{:.2f}%'.format((correctCount / len(y_test)) * 100))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
predictorPickleName = 'pickles/predictor-spacy-features.pkl'
dumpPickle(predictorPickleName, predictor)
```
